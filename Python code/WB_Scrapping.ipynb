```python
import wbgapi as wb
import pandas as pd
```

```python
pop_data = wb.data.DataFrame('SP.POP.TOTL',time=range(2000,2020))
```

```python
# Let's look at the result
pop_data
```

```python
# Moves years columns to rows
pop_data_melt = pd.melt(pop_data.reset_index(), id_vars = ['economy'], value_vars = pop_data.columns, var_name = "Year", value_name = "Value", ignore_index = False) 
```

```python
# Let's look at the result
pop_data_melt
```

```python
# Takes out the "YR" in front of Year value

year = []

for yryear in pop_data_melt["Year"]:
    year.append(int(yryear[2:]))

    
pop_data_melt["Year"] = year
```

```python
# Let's look at the result
pop_data_melt
```

```python
# Saves to CSV in Data folder in parent directory
pop_data_melt.to_csv('../Data/WB_Population.csv')
```

```python
# Scraps regions' codes and names

l_codes = []
l_names = []

for region in wb.region.list():
    l_codes.append(region["code"])
    l_names.append(region["name"])
    
d_regions = {"ID": l_codes, "Name": l_names}
df_regions = pd.DataFrame(d_regions)
df_regions.to_csv('../Data/WB_Regions.csv')
```

```python
# Scraps countries' codes, names, regions, and income levels

l_codes_countries = []
l_names_countries = []
l_regions_countries = []
l_incomelevels_countries = []

for country in wb.economy.list():
    if country["region"] != "":
        l_codes_countries.append(country["id"])
        l_names_countries.append(country["value"])
        l_regions_countries.append(country["region"])
        l_incomelevels_countries.append(country["incomeLevel"])
    
d_countries = {"ID": l_codes_countries, "Name": l_names_countries, "RegionID": l_regions_countries, "IncomeLevelID": l_incomelevels_countries}
df_countries = pd.DataFrame(d_countries)
df_countries.to_csv('../Data/WB_Countries.csv')
```

```python
# Scraps income levels' codes and names

l_codes_incomelevels = []
l_names_incomelevels = []

for incomelevel in wb.economy.list():
    if incomelevel["value"].find("income") > 0 and incomelevel["value"].find("excluding") == -1:
        l_codes_incomelevels.append(incomelevel["id"])
        l_names_incomelevels.append(incomelevel["value"])
    
d_incomelevels = {"IncomeLevelID": l_codes_incomelevels, "Name": l_names_incomelevels}
df_incomelevels = pd.DataFrame(d_incomelevels)
df_incomelevels.to_csv('../Data/WB_IncomeLevels.csv')
```
